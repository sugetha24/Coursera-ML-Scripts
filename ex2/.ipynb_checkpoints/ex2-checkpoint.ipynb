```python
#%matplotlib inline
from __future__ import division                                                 
import matplotlib.pyplot as plt                                                 
import scipy.optimize as op
import numpy as np                                                              
import pandas as pd 
```

```python
def sigmoid(z):
    g = np.zeros(len(z))
    g = 1 / (1 + np.exp(-z))
    return g
```

```python
def costFn(theta, X, y, m):
    h = sigmoid(X.dot(theta))
    J = 0
    J = 1 / m * np.sum((-(y * np.log(h))) - ((1-y) * np.log(1-h)))
    return J#.flatten()#, grad.flatten()
```

```python
def grad(theta, X, y, m):
    h = sigmoid(X.dot(theta))
    J = 1 / m * np.sum((-(y * np.log(h))) - ((1-y) * np.log(1-h)))
    #J = np.mean((-(y * np.log(h))) - ((1-y) * np.log(1-h)))
    gg = 1 / m * (X.T.dot(h-y))
    return gg.flatten()
```

```python
def plotData(X, y, f):
    plt1 = f.add_subplot(111)
    pos = np.where(y == 1)[0]
    neg = np.where(y == 0)[0]

    plt1.scatter(X[pos, 0], X[pos, 1], marker='x', color='r')
    plt1.scatter(X[neg, 0], X[neg, 1], marker='o', color='y')
    plt.xlabel('Exam1 score')
    plt.ylabel('Exam2 score')
    plt.legend(('Admitted', 'Not Admitted'), loc='lower left', fontsize=8, ncol=3, bbox_to_anchor=(0, -0.3))
```

```python
# ============================================== Plot data ==============================================  #
data = pd.read_csv('ex2data1.txt', sep=",", header=None)                        
data.columns = ['x1', 'x2', 'y']                                                       
x1 = data.iloc[:, 0].values[:, None]                                                     
x2 = data.iloc[:, 1].values[:, None]                                                    
y = data.iloc[:, 2].values[:, None]

# join x1 and x2 to make one array of X
X = np.concatenate((x1, x2), axis=1)

m, n = X.shape

plotData(X, y, plt.figure())

# ========================================= Compute cost and Gradient ================================ #
X = np.c_[np.ones(X.shape[0]), X]

# Initialize fitting parameters
initial_theta = np.zeros(n + 1)[:, None]

# J, G = costFn(initial_theta, X, y, m)
J = costFn(initial_theta, X, y, m)
G = grad(initial_theta, X, y, m)

print 'Cost at initial theta (zeros):', J
print 'Gradient at initial theta (zeros):'
print '\n'.join('%f' % value for value in G)

# ========================================== Advanced optimization ==================================== #

# need to flatten y inorder for minimize to work!
y1 = y.flatten()
Result = op.minimize(fun = costFn, 
                    x0 = initial_theta, 
                    args = (X, y1, m),
                    method = 'Nelder-Mead',
                    options={'disp': True})#,
                    #jac = grad) # does not need jacobian! it calculates that itself.

print 'Cost at theta found by fminunc:', Result.fun 
print 'Theta: '
print '\n'.join ('%f' % val for val in Result.x)
theta = Result.x


plot_x = np.array([min(X[:, 2])-2, max(X[:, 2]+2)])

plot_y = -1/theta[2] * ((theta[1]*plot_x) + theta[0])

f = plt.figure()
plt2 = f.add_subplot(111)
plt2.plot(plot_x, plot_y)
plotData(X[:, 1:3], y, f)
plt.xlabel('Exam1 score')
plt.ylabel('Exam2 score')
plt2.legend(('', 'Admitted', 'Not Admitted'), loc='lower left', fontsize=8, ncol=3, bbox_to_anchor=(0, -0.3))


# ========================================== Predict and Accuracies ==================================== #

prob = sigmoid(np.array([1, 45, 85]).dot(theta.reshape(3, 1)))
print 'For a student with scores 45 and 85, we predict an admission probability of %f' % prob
```
